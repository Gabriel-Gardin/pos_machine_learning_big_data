# Introdução ao Aprendizado de Máquina
<hr><h3>Modelagem Regressiva</h3>
<ul>
    <li>A regressão linear é uma das principais técnicas de estatística e Aprendizado de Máquina</li>
    <li><b>O que é regressão?</b>: "Técnica que busca o relacionamento entre as variáveis (atributos e targets). Em outras palavras, é um método que encontra uma função aproximada que mapeia as variáveis."</li>
    <li>Os targets, outputs ou respostas, são chamados <b>variáveis dependentes</b>.</li>
    <li>Já os inputs são chamados <b>variáveis independentes</b>.</li>
    <li>Tradicionalmente os valores de target são contínuos.</li>
</ul>

# Introdução ao Aprendizado de Máquina
<hr><h3>Modelagem Regressiva</h3>
<ul>
    <li><b>Quando precisamos usar uma regressão?</b>:</li>
    <ul>
        <li>Avaliar a relação entre as variáveis (influência entre os fenômenos);</li>
        <li>Previsão de valores</li>
    </ul>
    <li>A regressão pode ser linear, que estabelece uma relação linear entre a variável dependente e as independentes.</li>
    <li>Na regressão linear, a média dos erros é nula.</li>
    <li>No entanto, nem sempre o fenômeno tem o comportamento linear, assim podemos assumir funções Hiperbólicas, Exponenciais e Polinomiais .</li>
</ul>

<ul>
    <li><b>Problemas a serem enfrentados?</b>:</li>
    <ul>
        <li>Underfitting: O modelo (função encontrada) não foi capaz de capturar a dependência existente.</li>
        <li>Overfitting: O modelo aprendeu "os dados" e não necessariamente o problema.</li>
    </ul>

     <ul>
        <li>R<sup>2</sup>: Coeficiente de Correlação</li>
         <li> Mean Absolute Error (MAE): Média absoluta dos erros</li>
         <li> Mean Squared Error (MSE): Média quadrada dos erros</li>
         <li> Root Mean Squared Error (RMSE): Raiz da média quadrada dos erros</li>
    </ul>
</ul>

```python
import numpy as np
from sklearn.linear_model import LinearRegression

x = np.array([5, 15, 25, 35, 45, 55]).reshape((-1, 1))
y = np.array([5, 20, 14, 32, 22, 38])

model = LinearRegression()
model.fit(x, y)

#model.score retorna o coeficiente de determinação R2
r_sq = model.score(x, y)
print('R2:', r_sq)

print('Intercepto:', model.intercept_)

print('Inclinação:', model.coef_)
```

```python
import matplotlib.pyplot as plt

#linspace números distribuídos de maneira uniforme em um intervalo
x_new = np.linspace(0, 60, 100)
#determina y a partir do modelo linear criado anteriormente
y_new = model.predict(x_new[:, np.newaxis])

#Plotando
plt.figure(figsize=(4, 3))
ax = plt.axes()
ax.scatter(x, y)
ax.plot(x_new, y_new)

ax.set_xlabel('x')
ax.set_ylabel('y')

ax.axis('tight')

plt.show()
```

```python
#Exemplo Covid
import pandas as pd
import numpy as np
url = "https://drive.google.com/uc?id=1CyrX0o55_RH8IE2DI_sy18M-fFVdds1P"

# carregar o dataset
df = pd.read_csv(url, names=['numero', 'data', 'provincia', 'regiao', 'atualizacao'
                             ,'confirmado', "obito", "recuperado" ])

#df.provincia.unique()

df = df[df.regiao=="Brazil"]
df.head()
```

```python
df.describe()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.preprocessing import PolynomialFeatures

#a ideia é construir uma visualização do modelo com relação aos dados originais
#podemos, dessa forma, encontrar um grau que dá melhor resultado (pelo menos visual)

obitos = pd.DataFrame(pd.to_numeric(df['obito']).values[1:80])
print(obitos)

y = np.array(obitos[0])
x = obitos.index

#definindo modelo de regressão
#primeiro estabelece os coeficientes
#observar como a mudança do grau do polinômio muda o ajuste da regressão
coefficients = np.polyfit(x,y,3)
#depois constrói o polinomio em si
p = np.poly1d(coefficients)

#construindo a curva que representa os resultados da aplicação do modelo
new_x = np.linspace(x[0], x[-1])
new_y = p(new_x)

plt.plot(x, y, "o", new_x, new_y)

```

```python
from sklearn import linear_model
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline

import warnings
warnings.simplefilter('ignore', np.RankWarning)

#está pegando os elementos nos índices pares
x_train=x.values[::2]
#está pegando os elementos nos índices ímpares
x_test =x.values[1:][::2]
 
y_train= y[::2]
y_test = y[1:][::2]

output_results = []
#graus que serão usados
d_degree = np.arange(1,21)
fig = plt.figure(figsize=(16, 20))
for degree in d_degree:
    
    poly_fit = np.poly1d(np.polyfit(x_train,y_train,degree))
     
    predito=poly_fit(x_test)
    error=np.mean((predito-y_test)**2)**(1/2)

    output_results.append(error)

    fig.add_subplot(7, 3, degree) 
    plt.plot(x_test, predito, label="Ajustar com grau "+str(degree)+" (Erro:"+str(round(error,2))+")")
    plt.scatter(x_train, y_train, s=5, color="C1")
    plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=3, ncol=2, mode="expand", borderaxespad=0.)
 
```

```python
#Encontrando o Grau do polinômio
plt.figure(figsize=(16,4))

#imprime o erro
plt.plot(output_results[2:20]) #pulei o 1 para não distorcer o gráfico    
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3) #Criei um objeto kNN...podemos variar o k se quisermos
knn.fit(x_train.reshape(-1, 1), y_train) #Induzindo modelo (treino)
resultado = knn.predict(x_test.reshape(-1, 1)) # Prevendo o conjunto de teste

error=np.mean((resultado-y_test)**2)**(1/2)

plt.plot(x_test, resultado, label="(Erro:"+str(round(error,2))+")")
plt.scatter(x_train, y_train, s=3, color="C1")
plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=3, ncol=2, mode="expand", borderaxespad=0.)

#erro parecido com a função polinomial? Qual grau?
```

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier


# induzindo modelo SVM (SVR - Support Vector Regressor)
svr_lin = svm.SVR(kernel='linear').fit(x_train.reshape(-1, 1), y_train)
svr_poly = svm.SVR(kernel='poly').fit(x_train.reshape(-1, 1), y_train) #degree=5

resultado_lin = svr_lin.predict(x_test.reshape(-1, 1)) # Prevendo o conjunto de teste
error_lin=np.mean((resultado_lin-y_test)**2)**(1/2)

resultado_poly = svr_poly.predict(x_test.reshape(-1, 1)) # Prevendo o conjunto de teste
error_poly = np.mean((resultado_poly-y_test)**2)**(1/2)

# criando uma solução com RF
rf = RandomForestClassifier().fit(x_train.reshape(-1, 1), y_train)
resultado_rf = rf.predict(x_test.reshape(-1, 1))
error_rf = np.mean((resultado_rf-y_test)**2)**(1/2)

plt.figure(figsize=(14,6))
plt.plot(x_test, resultado_lin, label="SVM Linear (Erro:"+str(round(error_lin,2))+")", color="C2")
plt.plot(x_test, resultado_poly, label="SVM Polinomial (Erro:"+str(round(error_poly,2))+")", color="C3")
plt.plot(x_test, resultado, label="kNN (Erro:"+str(round(error,2))+")", color="C4")
plt.plot(x_test, resultado_rf, label="RF (Erro:"+str(round(error_rf,2))+")", color="C6")
plt.scatter(x_train, y_train, s=3, color="C1")
plt.legend()
```

```python
from sklearn.metrics import r2_score
#from sklearn.metrics import mean_squared_error

print("kNN: "+str(np.round(r2_score(y_test,resultado),2)))
print("SVM Linear: "+str(np.round(r2_score(y_test,resultado_lin),2)))
print("SVM Polinomial: "+str(np.round(r2_score(y_test,resultado_poly),2)))
print("RF: "+str(np.round(r2_score(y_test,resultado_rf),2)))
```
