### 1) Filtrando uma imagem digital


```python
import cv2
from google.colab.patches import cv2_imshow
from PIL import Image
from numpy import *
from pylab import *
import matplotlib.pyplot as plt
import matplotlib.image as mpimg 
from skimage import filters
```

```python
imagem = cv2.imread('atividade1.png')
```

#### Detectando Borda

```python
# Detector Sobel
sobelx = cv2.Sobel(src=imagem, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=3) # Detector aplicado ao eixo X (ksize = tamanho da máscara)
sobely = cv2.Sobel(src=imagem, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=3) # Detector aplicado ao eixo Y
sobelxy = cv2.Sobel(src=imagem, ddepth=cv2.CV_64F, dx=1, dy=1, ksize=3) # Detector aplicado aos eixos X e Y

#ddepth=cv2.CV_64F -> Garante que nossa imagem filtrada terá as mesmas dimensões da imagem original

plt.figure(figsize=(16, 8))

plt.subplot(1,4,1)
plt.imshow(imagem, cmap='gray')
plt.title('Imagem Original')

plt.subplot(1,4,2)
plt.imshow(sobelx, cmap = 'gray')
plt.title('Detector aplicado ao eixo X'), 

plt.subplot(1,4,3)
plt.imshow(sobely, cmap = 'gray')
plt.title('Detector aplicado aos eixo Y')

plt.subplot(1,4,4)
plt.imshow(sobelxy,cmap = 'gray')
plt.title('Detector aplicado aos eixos X e Y')
```

```python
#Detector Canny
canny = cv2.Canny(imagem,10,20) #Parâmetros: (Imagem, limiar alto de intensidade do gradiente, limiar baixo de intensidade do gradiente)

plt.figure(figsize=(16, 8))
plt.subplot(1,2,1)
plt.imshow(imagem, cmap = 'gray')
plt.title('Imagem Original'), 

plt.subplot(1,2,2)
plt.imshow(canny,cmap = 'gray')
plt.title('Bordas da Imagem')

plt.show()
```

#### Melhorando o Contraste

```python
imagem_contraste = cv2.imread('atividade5.jpg')
cv2_imshow(imagem_contraste) 
```

```python
#Convertendo para o canal Lab
lab = cv2.cvtColor(imagem_contraste, cv2.COLOR_BGR2LAB)
cv2_imshow(lab)
```

```python
#Recuperando cada canal

plt.figure(figsize=(16, 8))

plt.subplot(1,3,1)
plt.imshow(lab[:,:,0])
plt.title('Canal L')

plt.subplot(1,3,2)
plt.imshow(lab[:,:,1])
plt.title('Canal A')

plt.subplot(1,3,3)
plt.imshow(lab[:,:,2])
plt.title('Canal B')
```

```python
#Aplicando CLAHE no canal de cor
#CLAHE (Contrast Limited Adaptive Histogram Equalization)
#Realiza a equalização de histogramas a partir do tamanho do tileGridSize (bloco semelhante as máscaras), 8x8 como tamanho padrão
#clipLimit = valor de limiar
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
cl = clahe.apply(lab[:,:,0])
plt.imshow(cl, cmap="gray")
```

```python
#Remontando a imagem com o canal modificado
uniao_canais = cv2.merge((cl,lab[:,:,1],lab[:,:,2]))
plt.imshow(uniao_canais)
```

```python
# Convertendo a imagem de Lab para RGB
final = cv2.cvtColor(uniao_canais, cv2.COLOR_LAB2BGR)

plt.figure(figsize=(16, 8))

plt.subplot(1,2,1)
plt.imshow(imagem_contraste)
plt.title('Original')

plt.subplot(1,2,2)
plt.imshow(final)
plt.title('Processada')
```
