```python
#carregando os datasets 
beers = pd.read_csv('/content/beers.csv')
breweries = pd.read_csv('/content/breweries.csv')
```

```python
#apagando a coluna Unnamed porque já temos uma chave primária (que será a brewery_id)
beers = beers.drop(columns=['Unnamed: 0'])
beers
```

Fazendo uma exploração inicial no dataset breweries

```python
#carregando o segundo dataset
breweries
```

```python
#vendo uma lista com os nomes das colunas do dataset Breweries
print(list(breweries.columns.values))
```

```python
#renomeando a coluna Unnamed: 0 para brewery_id para fazer o merge
breweries.columns = ['brewery_id','brewery_name','city','state']
```

```python
breweries
```

Anáise do dataset resultante do merge entre os datasets beers e breweries

```python
#fazendo o merge entre os dois datasets
beerDf = pd.merge(beers, breweries, how = "outer")
```

```python
beerDf
```

```python
beerDf.head
```

```python
print(beerDf.columns.values)
```

```python
#arrumando os nomes das colunas
beerDf.columns = ['ABV', 'IBU', 'BeerID', 'BeerName', 'Style', 'BreweryID', 'Ounces', 'BreweryName', 'City', 'State']
```

```python
beerDf
```

Explorando as cervejas Witbier

```python
#vendo cervejas que são do estilo Witbier
WitiDf = beerDf[beerDf['Style'] == "Witbier"]
WitiDf
```

```python
#vendo a gradação alcólica (ABV) média das cervejas Witbier em porcentagem

print(WitiDf.ABV.mean()*100,'%',sep='')
```

```python
#estatística descritiva dos valores relacionados ä cerveja Witbier
WitiDf.describe()
```

Visualização dos dados de beerDf

```python
beerDf['BeerName'].value_counts()[:20].plot(kind = "bar")

plt.title("Número de cervejarias")

plt.show()
```

```python
#Plot de estilo de cerveja mais comuns
sns.set_theme(style = 'darkgrid')
beerDf['Style'].value_counts()[:15].plot(kind = "bar", color = "red")
plt.xticks(rotation=90)
plt.title("Estilo de cervejas mais comuns")
```

```python
#descbrindo as 10 mais populares
beerscount = beerDf['Style'].value_counts().rename_axis('Beers').reset_index(name='Total')
beerscount = beerscount.head(10)

#plot do gráfico pizza das 10 cervejas mais populares
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
labels = beerscount['Beers']
totals = beerscount['Total']
ax.pie(totals, labels = labels)
plt.title('10 cervejas mais populares')
plt.show()
```

```python
#Plot de istribuição de IBU
plt.hist("IBU", 11, data = beerDf, alpha = 0.8, color = "green")
plt.title("Distribuição de IBU")
```

```python
#Plot de istribuição de ABV
plt.hist("ABV", 11, data = beerDf, alpha = 0.8, color = "pink")
plt.title("Distribuição de ABV")
```

```python
#só pra contar as cervejarias por estado
states_count_df = beerDf['State'].value_counts().rename_axis('States').reset_index(name='Num Breweries')
states_count_df.head(5)
```

```python
fig, axes = plt.subplots(1,2, sharey=True, figsize=(10,5))
sns.set_theme(style="whitegrid")
fig.suptitle('Top 10 Estados')

#beerDf.columns = ['ABV', 'IBU', 'BeerID', 'BeerName', 'Style', 'BreweryID', 'Ounces', 'BreweryName', 'City', 'State']

# Estado com mais cervejarias
sns.barplot(ax=axes[0], x="States", y="Num Breweries", data=states_count_df.head(10))
axes[0].set_title("Estados com mais cervejarias")

# Estado com menos cervejarias
sns.barplot(ax=axes[1], x="States", y="Num Breweries", data=states_count_df.tail(10))
axes[1].set_title("Estados com menos cervejarias")
```
