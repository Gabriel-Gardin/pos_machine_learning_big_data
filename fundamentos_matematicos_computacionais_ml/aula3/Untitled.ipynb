```python
import pandas as pd
import numpy as np
```

```python
df_beers = pd.read_csv('archive/beers.csv')
df_brew = pd.read_csv('archive/breweries.csv')
```

```python
df_brew.head()
```

```python
#Renomeia as colunas do pandas.

df_brew.columns = ['brewery_id', 'name', 'city', 'state']
```

```python
df_brew.head()
```

```python
df_beers.head()
```

```python
#Junta os dois data frames aonde o mesmo brewery id é identificado
df = pd.merge(df_beers, df_brew, on=['brewery_id'])
df.head()
```

```python
#Remove a coluna sem nome
df = df.drop(['Unnamed: 0'], axis=1)
df.head()
df.columns = ['abv', 'ibu', 'id', 'name_beer', 'estilo', 'brewery_id', 'ounces', 'brewery_name', 'city', 'state']
```

```python
df
```

```python
#Qaul estado produz mais cervejas?
#Resposta: CO

df.state.value_counts() #Retorna a frequencia com que cada ítem aparece no DF.
```

```python
#Qual o tipo de cerveja mais produzida?

#Resposta: American IPA

df.estilo.value_counts() #Retorna a frequencia com que cada ítem aparece no DF.
```

```python
#Qual fabricante tem mais estilos de cervejas?

#Resposta: Brewery Vivant

df.brewery_name.mode()
```

```python
#Quais valores exclusivos de tipos de cervejas?

#Resposta 100 cervejas exclusivas.

a = df.estilo.unique()
len(a)
```

```python
#Quais cervejas produzidas em Denver(CO)

a = df.loc[df['state'] == ' CO']['name_beer']
a.value_counts()
```
