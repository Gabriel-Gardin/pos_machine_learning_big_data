## Muitos dados foram preenchidos como zero, portanto serão considerados como ausentes.

Os atributos nulos que possuem distribuição normal serão substituídos pela média, 
enquanto que os não normais serão substituídos pela mediana.

Outra opção para imputar dados seria usar https://github.com/YA26/MissingValuesHandler que faz a imputação de dados usando um algoritmo de floresta aleatória

```python
df2 = df[df["Glucose"] != 0]
df2.Glucose.kurtosis()

#Distribuição normal
```

```python
#Como todos os parametros seguem uma distribuição proxima da normal vamos usar a média para imputar valores faltantes

df['Glucose'].replace(0,df['Glucose'].mean(),inplace=True)
df['BloodPressure'].replace(0,df['BloodPressure'].mean(),inplace=True)
df['SkinThickness'].replace(0,df['SkinThickness'].mean(),inplace=True)
df['BMI'].replace(0,df['BMI'].mean(),inplace=True)
df['Insulin'].replace(0,df['Insulin'].mean(),inplace=True)
```

```python
from imblearn.over_sampling import SMOTE

os = SMOTE(random_state = 1)
os_data_x, os_data_y = os.fit_resample(x_train, y_train)
os_data_x= pd.DataFrame(os_data_x, columns=x_train.columns)
os_data_y= pd.DataFrame(os_data_y, columns=['Outcome'])
#Geração de dados sinteticos para balanceamento
os_data_y.value_counts()
```

## Escalonamento das características

```python
from sklearn.preprocessing import StandardScaler

StdSc=StandardScaler()

for col in os_data_x.columns:
    os_data_x[col]=StdSc.fit_transform(os_data_x[[col]])
    

#Escalonando o dataset de validação

for col in x_test.columns:
    x_test[col]=StdSc.fit_transform(x_test[[col]])
```

## Treinando os modelos

```python
from sklearn.linear_model import LogisticRegression

logreg= LogisticRegression()
logreg.fit(os_data_x, os_data_y.values.ravel())
```

## Validação cruzada, método k-fold

```python
cross_val_score(logreg, os_data_x, os_data_y, cv=3, scoring="accuracy")
```

```python
y_pred=logreg.predict(x_test)
```

```python
#Acurácia com os dados de teste
logreg.score(x_test, y_test)
```

```python
def plot_confusion_matrix(matrix):
    fig = plt.figure(figsize=(8,8))
    ax = fig.add_subplot(111)
    cax = ax.matshow(matrix)
    fig.colorbar(cax)
```

```python
from sklearn.metrics import confusion_matrix

conf_mx = confusion_matrix(y_test, y_pred)
conf_mx
```

```python
plot_confusion_matrix(conf_mx)
```

```python
def plot_roc_curve(fpr, tpr, label=None):
    plt.plot(fpr, tpr, linewidth=2, label=label)
    plt.plot([0, 1], [0, 1], 'k--') # dashed diagonal
    plt.axis([0, 1, 0, 1])                                   
    plt.xlabel('False Positive Rate (Fall-Out)', fontsize=16) 
    plt.ylabel('True Positive Rate (Recall)', fontsize=16)    
    plt.grid(True) 
```

## Vamos utilizar o método cross_val_predict para fazer o gráfico da curva ROC sem usar os dados de teste

```python
from sklearn.model_selection import cross_val_predict


y_scores = cross_val_predict(logreg, os_data_x, os_data_y.values.ravel(), cv=3,
                             method="decision_function")
```

```python
fpr, tpr, thresholds = roc_curve(os_data_y.values.ravel(), y_scores)
```

```python
plt.figure(figsize=(8, 6))                               
plot_roc_curve(fpr, tpr)
#save_fig("roc_curve_plot")                                    
plt.show()
```

```python
roc_auc_score(os_data_y.values.ravel(), y_scores)
```
