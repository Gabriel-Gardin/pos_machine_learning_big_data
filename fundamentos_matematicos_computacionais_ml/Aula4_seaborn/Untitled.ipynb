```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('archive/IRIS.csv')
df.head(5)
```

Histograma com seaborn:

```python
sns.histplot(df.sepal_length)
plt.xlabel('Sepal length')
plt.ylabel('Contagem')
plt.title('Histograma - Sepal length')
plt.show()
```

Point plot

```python
sns.pointplot(x=df.sepal_length, y=df.sepal_width)
sns.pointplot(x=df.petal_length, y=df.petal_width)
plt.show()
```

```python
sns.jointplot(x=df.sepal_length, y=df.sepal_width, hue=df.species)
```

```python
sns.jointplot(x=df.sepal_length, y=df.sepal_width, hue=df.species, kind='kde')
```

PIECHART

```python
plt.pie(df.species.value_counts(), labels=df.species.value_counts().index)
plt.show()
```
